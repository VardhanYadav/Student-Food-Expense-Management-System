```python
import pandas as pd

# Load the dataset
from google.colab import files
uploaded = files.upload()
```

```python
# Replace 'food_data.csv' with the actual filename if different
df = pd.read_csv('food_coded (1).csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=df, x='Gender', y='GPA', palette='pastel')
plt.title('GPA Distribution by Gender')
plt.show()
```

```python
sns.histplot(data=df, x='calories_day', kde=True, bins=30, color='orange')
plt.title('Distribution of Daily Calorie Intake')
plt.xlabel('Calories per Day')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(data=df, x='healthy_feeling', y='GPA', hue='Gender')
plt.title('Healthy Feeling vs GPA')
plt.xlabel('Healthy Feeling (Scale)')
plt.ylabel('GPA')
plt.show()
```

```python
sns.barplot(data=df, x='employment', y='eating_out', hue='Gender', palette='viridis')
plt.title('Eating Out Frequency by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Eating Out Frequency')
plt.show()
```

```python
fav_cuisine_counts = df['fav_cuisine'].value_counts()
fav_cuisine_counts.plot(kind='bar', color='purple')
plt.title('Favorite Cuisine Distribution')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
exercise_gender = df.groupby(['Gender', 'exercise']).size().unstack()
exercise_gender.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.title('Exercise Habits by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
nutritional_heatmap = df.groupby(['nutritional_check', 'healthy_meal']).size().unstack()
# Change the format code to '.0f' or 'g' to handle floats
sns.heatmap(nutritional_heatmap, annot=True, fmt='.0f', cmap='YlGnBu')
plt.title('Nutritional Check vs Healthy Meal')
plt.xlabel('Healthy Meal')
plt.ylabel('Nutritional Check')
plt.show()
```

```python
sns.scatterplot(data=df, x='income', y='calories_day', hue='Gender')
plt.title('Income vs Daily Calorie Intake')
plt.xlabel('Income')
plt.ylabel('Calories per Day')
plt.show()
```

```python
weight_perception = df['self_perception_weight'].value_counts()
weight_perception.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'pink'])
plt.title('Self-Perception of Weight')
plt.ylabel('')
plt.show()
```

```python
df[['veggies_day', 'fruit_day']].mean().plot(kind='bar', color=['green', 'orange'])
plt.title('Average Veggies and Fruits Consumed per Day')
plt.ylabel('Average Servings')
plt.show()
```

```python
import plotly.express as px

fig = px.scatter(df, x='calories_day', y='GPA', color='Gender', title='Calories per Day vs GPA')
fig.show()
```
